```python
import numpy as np
from bokeh.io import output_notebook
from bokeh.plotting import figure, output_file, show
from bokeh.layouts import row, column
from bokeh.models.sources import ColumnDataSource

import time

output_notebook()
```

```python
# data
em = ColumnDataSource(data=dict(x=[0], y=[0]))
ec = ColumnDataSource(data=dict(x=[0], y=[0]))

# EM
em_fig = figure(plot_width=800, plot_height=250)
em_fig.line("x", "y", source=em, legend='EM')

ec_fig = figure(plot_width=800, plot_height=250)
ec_fig.line("x", "y", source=ec, line_color='red', legend='EC')

handle = show(column(em_fig, ec_fig), notebook_handle=True)

new_data_em = dict(x=[0], y=[0])
new_data_ec = dict(x=[0], y=[0])
y_em = []
y_ec = []
x = []

step = 0
step_size = 1 # increment for increasing step
max_step = 100000  # arbitrary stop point for example
period = .1  # in seconds (simulate waiting for new data)
n_show = 10  # number of points to keep and show

f = open('running_avg.dat', 'r')
f.seek(0,2)
while step < max_step:
    newline = f.readline()
    if newline:
        a0 = list(map(np.float, newline.split()))
        x.append(step)
        y_em.append(a0[0])
        y_ec.append(a0[1])
        new_data_em['x'] = x = x[-n_show:]  # prevent filling ram
        new_data_ec['x'] = x = x[-n_show:]  # prevent filling ram
        new_data_em['y'] = y_em = y_em[-n_show:]  # prevent filling ram
        new_data_ec['y'] = y_ec = y_ec[-n_show:]  # prevent filling ram

        em.stream(new_data_em, n_show)
        ec.stream(new_data_ec, n_show)

        push_notebook(handle=handle)
        step += step_size
    time.sleep(period)
```
